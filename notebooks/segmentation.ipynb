```python
#imports

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

```python
# load the data

data = pd.read_csv('../data/Mall_Customers.csv')
```

```python
scaler = StandardScaler()
data[['Age T','Annual Income T','Spending Score T']] = scaler.fit_transform(data[['Age','Annual Income (k$)','Spending Score (1-100)']])
```

```python
def optimise_k_means(data, max_k):
    means = []
    inertias = []

    for k in range (1, max_k):
        kmeans = KMeans(n_clusters=k)
        kmeans.fit(data)

        means.append(k)
        inertias.append(kmeans.inertia_)

    fig = plt.subplots(figsize=(10,5))
    plt.plot(means, inertias, 'o-')
    plt.xlabel("Number of clusters")
    plt.ylabel("Inertia")
    plt.grid(True)
    plt.show()
```

```python
optimise_k_means(data[['Age T', 'Annual Income T','Spending Score T']],10)
```

```python
model = KMeans(n_clusters = 4)
```

```python
model.fit(data[['Age T', 'Annual Income T','Spending Score T']])
```

```python
data['cluster'] = model.labels_
```

```python
data.head()
```

```python
data_grouped = data.groupby(['cluster','Gender'])['CustomerID'].count()
```

```python
fig = plt.figure(figsize=(12,12))
ax = fig.add_subplot(projection='3d')

ax.scatter(
    data['Age'],
    data['Annual Income (k$)'],
    data['Spending Score (1-100)'],
    c = data['cluster'],
    cmap='Set1',
    s=50)

ax.set_xlabel('Age')
ax.set_ylabel('Annual Income (k$')
ax.set_zlabel('Spending Score (1-100)')

ax.view_init(20,-35)

plt.show()
```
